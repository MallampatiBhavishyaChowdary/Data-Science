```python
corr = auto_df.corr()
```

```python
# Set the figure size
plt.figure(figsize =(8,6))

# Create heatmap of passengers by class
sns.heatmap(corr, annot = True)

# Set the title and axis labels
plt.title('Correlation Matrix')
plt.xlabel('Features')
plt.ylabel('Features')

# Show the plot
plt.show()
```

```python
import plotly.graph_objects as go
explode_effect =[0.1,0.1,0.1]
custom_colors = ['Deeppink', 'aqua', 'orange']

fig= go.Figure(go.Pie(
      labels = auto_df["cylinders"],
      
      pull=explode_effect,
      marker=dict(colors=custom_colors)
))


# Set the title
fig.update_layout(title="Cylinders Distribution")

# Show the pie chart
fig.show()
```
